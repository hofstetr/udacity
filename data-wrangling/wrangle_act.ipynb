# Clean

```python
# Make copies of the dataframes for cleaning
archive_clean = archive.copy()
breeds_clean = breeds.copy()
```

#### 'timestamp' incorrect data type
##### Define
Convert column to datetime
##### Code

```python
archive_clean['timestamp'] = pd.to_datetime(archive_clean['timestamp'])
```

##### Test

```python
archive_clean['timestamp'].head()
```

#### Records with incorrect ratings
##### Define
Omit 23 records where the ratings_denominator is not 10
##### Code

```python
archive_clean = archive_clean[archive_clean['rating_denominator'] == 10]
```

##### Test

```python
archive_clean[archive_clean['rating_denominator'] != 10]['tweet_id'].count()
```

#### Records with invalid dog names
##### Define
Omit 95 records with invalid dog names
##### Code

```python
archive_clean = archive_clean[~archive_clean['name'].isin(invalid_names)]
```

##### Test

```python
archive_clean[archive_clean['name'].isin(invalid_names)]
```

#### Archive includes retweets
##### Define
Omit 181 retweet records and drop the retweet information entirely
##### Code

```python
archive_clean = archive_clean[archive_clean['retweeted_status_id'].isnull()]
archive_clean.drop(columns=['retweeted_status_id','retweeted_status_user_id', 'retweeted_status_timestamp'], inplace=True)
```

##### Test

```python
archive_clean.info()
```

#### Two columns have very little data to be of use
##### Define
Drop these two columns from the cleaned dataframe
##### Code

```python
archive_clean.drop(columns=['in_reply_to_status_id', 'in_reply_to_user_id'], inplace=True)
```

##### Test

```python
archive_clean.columns
```

#### Inconsistency in breed names
##### Define
Change all to lower case and remove extraneous separators
##### Code

```python
breeds_clean['p1'] = breeds_clean['p1'].str.lower().str.replace(r'[_|-]',' ')
breeds_clean['p2'] = breeds_clean['p2'].str.lower().str.replace(r'[_|-]',' ')
breeds_clean['p3'] = breeds_clean['p3'].str.lower().str.replace(r'[_|-]',' ')
```

##### Test

```python
breeds_clean.head()
```

# Tidiness

#### Important counts necessary to perform analysis in separate file
##### Define
Add retweet_count and favorite_count to the archive dataframe
##### Code

```python
# Add retweet_count and favorite_count to the archive data set
archive_clean['retweet_count'] = 0
archive_clean['favorite_count'] = 0
for id in archive_clean['tweet_id']:
    if id in tweets.keys():
        archive_clean.loc[archive['tweet_id'] == id, 'retweet_count'] = tweets[id]['retweet_count']
        archive_clean.loc[archive['tweet_id'] == id, 'favorite_count'] = tweets[id]['favorite_count']
```

```python
##### Test
```

```python
archive_clean[['tweet_id','retweet_count','favorite_count']].head()
```

#### Dog breed predictions are separate from rest of tweet data
##### Define
Separate into three sets of predictions then concatenate them and filter on the row that had the highest confidence
##### Code

```python
# Separate into three datasets
preds1 = breeds_clean[['tweet_id','jpg_url','img_num','p1','p1_conf','p1_dog']]
preds2 = breeds_clean[['tweet_id','jpg_url','img_num','p2','p2_conf','p2_dog']]
preds3 = breeds_clean[['tweet_id','jpg_url','img_num','p3','p3_conf','p3_dog']]
```

```python
# Rename the columns
preds1 = preds1.rename(columns={'p1':'breed', 'p1_conf':'confidence', 'p1_dog':'dog'})
preds2 = preds2.rename(columns={'p2':'breed', 'p2_conf':'confidence', 'p2_dog':'dog'})
preds3 = preds3.rename(columns={'p3':'breed', 'p3_conf':'confidence', 'p3_dog':'dog'})
```

```python
# Concatenate them and filter for highest confidence as well as only dog images
breeds_clean = pd.concat([preds1,preds2,preds3])
breeds_clean = breeds_clean.sort_values('confidence', ascending=False).drop_duplicates('tweet_id')
breeds_clean = breeds_clean[breeds_clean['dog'] == True]
breeds_clean
```

```python
# Finally we can join the breed onto the archive
archive_clean = pd.merge(archive_clean, breeds_clean[['tweet_id','breed']], on=['tweet_id'])
```
