## Investigation Overview

> In this notebook I will show one way to predict flight arrival delays to help carriers make recommendations to consumers about how much time to allow for connecting flights when they are in the booking process.

## Dataset Overview

> The data for this analysis was obtained from the [Bureau of Transportation Staticis](https://www.transtats.bts.gov/DL_SelectFields.asp?Table_ID=236). All domestic United States regions for years 2018 and 2019 were download by month and concatenated together outside of this notebook into two separate files for each year. It is a rather large set of data so I've separated the cleaning in order to speed processing of cells.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
delayed_flights = pd.read_csv('delayed_flights.csv')
# create a dictionary of carriers for plot labels
carriers = {'9E':'Pinnacle Airlines Inc.',
            'AA':'American Airlines Inc.',
            'AS':'Alaska Airlines Inc.',
            'B6':'JetBlue Airways',
            'DL':'Delta Air Lines Inc.',
            'EV':'Atlantic Southeast Airlines',
            'F9':'Frontier Airlines Inc.',
            'G4':'Allegiant Air',
            'HA':'Hawaiian Airlines Inc.',
            'MQ':'American Eagle Airlines Inc.',
            'NK':'Spirit Air Lines',
            'OH':'Comair Inc.',
            'OO':'Skywest Airlines Inc.',
            'UA':'United Air Lines Inc.',
            'VX':'Virgin America',
            'WN':'Southwest Airlines Co.',
            'YV':'Mesa Airlines Inc.',
            'YX':'Midwest Airline, Inc.'}
```

## Average Delays

> Note that when an arrival delay does occur there are none less than 15 minutes and the mean seems to be 30 minutes with extremes as high as 1,920 minutes.

```python
# Normalize the arrival delays
delayed_arrivals = delayed_flights[delayed_flights['ARR_DELAY_MEAN'] > 0].copy()
delayed_arrivals['ARR_DELAY_NORM'] = np.log(delayed_arrivals['ARR_DELAY_MEAN'])

# Generate a common tick scale
delay_ticks = [1,5,15,30,60,120,240,480,960,1920]
log_ticks = np.log(delay_ticks)

# Plot a histogram of the normalized data
bins = np.arange(0, delayed_arrivals['ARR_DELAY_NORM'].max()+.4, .4)
plt.figure(figsize=[12,5])
sb.distplot(delayed_arrivals['ARR_DELAY_NORM'], bins=bins)
plt.xticks(log_ticks, delay_ticks);
plt.xlabel('delay (minutes)');
plt.title('Average Arrival Delays');
```

## Number of Trips

> It turns out the number of trips made between cities is a factor in predicting the anticipated arrival delay when a delay does occur. When it is a major trip with a single daily flight the mean delay is around 30 minutes whereas the mean for very frequent trips is as high as 60 minutes.

```python
# Generate a new tick scale for the next plot since arrival delays start at 15 minutes
delay_ticks = [10,15,30,60,120,240,480,960,1920]
log_ticks = np.log(delay_ticks)

default_color = sb.color_palette()[0]
plt.figure(figsize=[12,5])
sb.boxplot(data=delayed_arrivals, x='FLIGHTS', y='ARR_DELAY_NORM', color=default_color);
plt.xlabel('number of flights');
plt.ylabel('average delay (minutes)');
plt.yticks(log_ticks, delay_ticks);
plt.title('Arrival Delays by Number of Trips');
```

## Comparing Delays and Trips across Carriers

> Many carriers follow the pattern of more trips equates to longer delays but there are outliers. JetBlue can anticpate shorter delays when a segment makes ten trips and Delta can anticipate the longest delays when they operate a segment that makes eight trips.

```python
# Generate a new tick scale for this plot
delay_ticks = [10,15,30,60,120]
log_ticks = np.log(delay_ticks)

flights = delayed_arrivals['FLIGHTS'].unique()
g = sb.FacetGrid(data=delayed_arrivals, col='OP_UNIQUE_CARRIER', col_wrap=5);
g.map(sb.barplot, 'FLIGHTS', 'ARR_DELAY_NORM', order=flights);
g.set_axis_labels('number of flights', 'delay (minutes)')
g.set(yticks=log_ticks);
g.set_yticklabels(delay_ticks);
for ax, title in zip(g.axes.flat, carriers.values()):
    ax.set_title(title)
```
